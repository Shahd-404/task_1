```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
import os

audio = AudioSegment.from_wav(RAW_WAV_PATH)

chunks = split_on_silence(
    audio,
    min_silence_len=350,
    silence_thresh=audio.dBFS - 20,
    keep_silence=200
)

# مسح القديم
for f in os.listdir(CHUNKS_DIR):
    os.remove(os.path.join(CHUNKS_DIR, f))

# حفظ الجديد
for i, chunk in enumerate(chunks):
    out_path = os.path.join(CHUNKS_DIR, f"audio_{i:04d}.wav")
    chunk.export(out_path, format="wav")

print("Detected chunks:", len(chunks))
print("Expected lines :", len(script_lines))
```

```python
import os
import librosa

chunk_files = sorted([f for f in os.listdir(CHUNKS_DIR) if f.endswith(".wav")])

print("First 10 files:")
for f in chunk_files[:10]:
    path = os.path.join(CHUNKS_DIR, f)
    y, sr = librosa.load(path, sr=None)
    duration = len(y) / sr
    print(f, "->", round(duration, 2), "sec")
```

```python
from datasets import Dataset, Audio
import pandas as pd
import os

df = pd.read_csv(METADATA_PATH)

df["audio"] = df["file"].apply(lambda x: os.path.join(CHUNKS_DIR, x))
df = df[["audio", "text"]]

my_voice_dataset = Dataset.from_pandas(df)
my_voice_dataset = my_voice_dataset.cast_column("audio", Audio(sampling_rate=16000))

print(my_voice_dataset)
print(my_voice_dataset[0])
```

```python
from datasets import DatasetDict

split_ds = my_voice_dataset.train_test_split(test_size=0.15, seed=42)

voice_dataset = DatasetDict({
    "train": split_ds["train"],
    "validation": split_ds["test"]
})

print(voice_dataset)
print("train:", len(voice_dataset["train"]))
print("validation:", len(voice_dataset["validation"]))
```

```python
VOICE_DATASET_DIR = os.path.join(BASE_DIR, "hf_voice_dataset")
voice_dataset.save_to_disk(VOICE_DATASET_DIR)

print("Saved dataset to:", VOICE_DATASET_DIR)
```
